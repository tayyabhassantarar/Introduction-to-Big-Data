```python
 #  Steps for data cleaning


import pandas as pd;     # 1- Import Libraries
import numpy as np;
df=pd.read_csv('auto.csv')   #2- pd.read_csv file for data manipulations 

headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]    
df.columns=headers                        # 3- Replace Headers
df.replace("?",np.nan, inplace=True)      # 4- replace ? places with NaN
df.isnull().sum()                         # 5- Check headers where NaN exists
df['normalized-losses'].head(25)          # 6- Check datatype for specified header

m=df['normalized-losses'].astype('float').mean()          # 7- find mean(),min(),max(),value_counts,.. by cosidering .astype
df['normalized-losses'].replace(np.nan,m,inplace =True)   # 8- Replace find out functions values at NaN places of header
df.isnull().sum()                                         # 9- check the NaN values for the header
```

```python
df['num-of-doors'].head(25)
maxi=df['num-of-doors'].value_counts().idxmax()
df['num-of-doors'].replace(np.nan,maxi,inplace=True)
df    
```

```python
df['bore'].head(25)
BoreMean=df['bore'].astype('float').mean()
df['bore'].replace(np.nan, BoreMean, inplace=True )
df
```

```python
df.isnull().sum()
```

```python
df['stroke'].head(25)
StrokeMean= df['stroke'].astype(float).mean()
df['stroke'].replace (np.nan, StrokeMean, inplace =True)
df
```

```python
df.isnull().sum()
```

```python
df['horsepower'].head(25)
hp=df['horsepower'].value_counts().idxmax()
df['horsepower'].replace(np.nan, hp, inplace=True)
df
```

```python
df.isnull().sum()
```

```python
df['peak-rpm'].head(25)
Prpm= df['peak-rpm'].value_counts().idxmax()
df['peak-rpm'].replace(np.nan, Prpm, inplace=True )
df
```

```python
df.isnull().sum()
```

```python
df['price'].head(25)
pr= df['price'].value_counts().idxmax()
df['price'].replace(np.nan, pr, inplace =True )
df
```
