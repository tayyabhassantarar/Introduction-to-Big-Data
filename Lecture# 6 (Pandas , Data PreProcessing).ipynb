```python
# make the whole changes in the file on question marks
import numpy as np
```

```python
df.replace("?",np.nan,inplace=True)   # it replaces the ? to nam that means there would be a null , and it displays NaN that place
df
```

```python
df.isnull()   # place false at the places where data exists, and true at where there exists no data
```

```python
df.isnull().sum()   # tells how much there is a null data, or how much there are true values
```

```python
# Now how to replaced data on these null places?
df['normalized-losses'].head(25)    # shows the type of object
```

```python
df['normalized-losses'].astype ('float').mean() # Find the mean (average)
```

```python
m=df ['normalized-losses'].astype('float').mean()       
df['normalized-losses'].replace(np.nan,m,inplace=True)
df.isnull().sum()
```
